```python
def visualize(orignal,augmented):
    fig = plt.figure()
    plt.subplot(1,2,1)
    plt.title('Original')
    plt.imshow(orignal)
    plt.subplot(1,2,2)
    plt.title('Augmented')
    plt.imshow(augmented)
```

```python
flipped = tf.image.flip_left_right(image)
```

```python
plt.imshow(flipped)
```

```python
visualize(image,flipped)
```

```python
grayscaled = tf.image.rgb_to_grayscale(image)
```

```python
visualize(image,tf.squeeze(grayscaled))
```

```python
plt.imshow(tf.squeeze(grayscaled),cmap='gray_r')
```

```python
saturated = tf.image.adjust_saturation(image,8)
visualize(image,saturated)
```

```python
bright = tf.image.adjust_brightness(image,0.3)
visualize(image,bright)
```

```python
rotated = tf.image.rot90(image)
visualize(image,rotated)
```

```python
cencropped = tf.image.central_crop(image,central_fraction=0.5)
visualize(image,cencropped)
```

```python
contras = tf.image.adjust_contrast(image,0.3)
visualize(image,contras)
```

```python
aug1 = tf.image.flip_up_down(image)
visualize(image,aug1)
```
