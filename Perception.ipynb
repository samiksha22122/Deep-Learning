```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
# x
independent = np.array([-1,0,1,2,3,4]).reshape(-1,1)
```

```python
# y
dependent = np.array([-3,-1,1,3,5,7])
```

```python
plt.scatter(independent,dependent)
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(independent,dependent)
```

```python
#b0
lin_reg.intercept_
```

```python
#b1
lin_reg.coef_
```

```python
plt.scatter(independent,dependent)
plt.plot(independent,lin_reg.predict(independent),c='red')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
# Preceptron
#Keras is a high-level, deep learning API developed by Google for implementing neural networks
#It is written in Python and is used to make the implementation of neural networks easy.
#It also supports multiple backend neural network computation.

#Sequential. The core idea of Sequential API is simply arranging the Keras layers in a sequential order and so,
#it is called Sequential API (application programming interface)
#Most of the ANN also has layers in sequential order and the data flows from one layer to another layer in the 
#,given order until the data finally reaches the output layer.

#Dense layer is the regular deeply connected neural network layer. 
#It is most common and frequently used layer
# Dense layer does the below operation on the input and return the output. output = activation(dot(input, kernel) + bias)
```

```python
import keras
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential([Dense(units=1,input_shape=[1])])
```

```python
model.summary()
```

```python
model.compile(loss='mean_squared_error',optimizer='SGD')
```

```python
model.fit(independent,dependent,epochs=500)
```

```python
model.get_weights() #Returns the current weights of the layer, as NumPy arrays. 
#The weights of a layer represent the state of the layer
# This function returns a list consisting of NumPy arrays. 
#The first array gives the weights of the layer and the second array gives the biases.
```

```python
plt.scatter(independent,dependent)
plt.plot(independent,model.predict(independent),c='red')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```
