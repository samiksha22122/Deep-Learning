```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set_style('whitegrid')
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
import tensorflow as tf
```

```python
from sklearn.datasets import make_regression
```

```python
X,y = make_regression(n_samples=1000,n_features=1,noise=60,random_state=1)
```

```python
X
```

```python
y
```

```python
plt.scatter(X,y)
plt.show()
```

```python
model = Sequential([
    Dense(10,activation='relu',input_shape=[1],kernel_regularizer=keras.regularizers.L2(10)), # hidden_layer1
    Dense(10,activation='relu'), #hidden_layer2
    Dense(10,activation='relu'), # hidden_layer3
    Dense(1)
])
```

```python
model.summary()
```

```python
optimizer = tf.keras.optimizers.RMSprop(learning_rate=0.01)
model.compile(loss='mean_squared_error',optimizer=optimizer)
```

```python
model.fit(X,y,epochs=500)
```

```python
plt.scatter(X,y)
plt.plot(X,model.predict(X),c='red')
plt.show()
```
