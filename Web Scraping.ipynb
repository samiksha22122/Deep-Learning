```python
import bs4 as bs
import urllib.request
```

```python
source = urllib.request.urlopen('https://en.wikipedia.org/wiki/Carbon_footprint').read()
```

```python
source
```

```python
soup = bs.BeautifulSoup(source,'lxml')
```

```python
soup
```

```python
text = ""
for paragraph in soup.find_all('p'):
    text += paragraph.text
```

```python
text
```

```python
import re
text = re.sub(r'\[[0-9]*\]', ' ',text)
text = re.sub(r'\s+',' ',text)
clean_text = text.lower()
clean_text = re.sub(r'\W',' ',clean_text)
clean_text = re.sub(r'\d',' ',clean_text)
clean_text = re.sub(r'\s+[a-z]\s+',' ',clean_text)
clean_text = re.sub(r'\s+',' ',clean_text)
```

```python
text
```

```python
clean_text
```

```python
import nltk
nltk.download('punkt')
```

```python
sentences = nltk.sent_tokenize(text)
```

```python
sentences
```

```python
len(sentences)
```

```python
stop_words = nltk.corpus.stopwords.words('english')
```

```python
stop_words 
```

```python
word2count = {}
for word in nltk.word_tokenize(clean_text):
    if word not in stop_words:
        if word not in word2count.keys():
            word2count[word] = 1
        else:
            word2count[word] += 1
```

```python
word2count
```

```python
for key in word2count.keys():
    word2count[key] = word2count[key]/(max(word2count.values())) #139/139
```

```python
word2count
```

```python
sent2score = {}
for sentence in sentences:
    for word in nltk.word_tokenize(sentence.lower()):
        if word in word2count.keys():
            if len(sentence.split(' '))<50:
                if sentence not in sent2score.keys():
                    sent2score[sentence] = word2count[word]
                else:
                    sent2score[sentence] += word2count[word]
```

```python
sent2score
```

```python
import heapq
```

```python
best_sentences = heapq.nlargest(20,sent2score,key=sent2score.get)
```

```python
best_sentences
```

```python
best_sentences = ' '.join(best_sentences)
```
