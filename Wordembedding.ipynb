```python
import nltk
import urllib
import bs4 as bs
import re
from gensim.models import Word2Vec
```

```python
sources = urllib.request.urlopen('https://en.wikipedia.org/wiki/Sachin_Tendulkar').read()
```

```python
sources
```

```python
soup = bs.BeautifulSoup(sources,'lxml')
```

```python
soup
```

```python
text = ""
for paragraph in soup.find_all('p'):
  text += paragraph.text
```

```python
text
```

```python
text = re.sub(r'\[[0-9]*\]',' ',text)
text = text.lower()
text = re.sub(r'[@#\$%&\*\(\)\<\>\?\'\":;\-\,]',' ',text)
text = re.sub(r'\d',' ',text)
text = re.sub(r'\s+',' ',text)
```

```python
text
```

```python
nltk.download('punkt')
sentences = nltk.sent_tokenize(text)
```

```python
len(sentences)
```

```python
sentences
```

```python
sentences = [nltk.word_tokenize(sentence) for sentence in sentences]
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
for i in range(len(sentences)):
  sentences[i] = [word for word in sentences[i] if word not in stopwords.words('english')]
```

```python
sentences
```

```python
model = Word2Vec(sentences,min_count=1)
```

```python
model.vector_size
```

```python
model.wv.vocab
```

```python
vector = model.wv['srikkanth']
```

```python
vector
```
